```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
nadia = cv2.imread('E:/Images/nikola_tesla.png',0)
denis = cv2.imread('E:/Images/solvay_conference.jpg',0)
solvay = cv2.imread('E:/Images/albert_einstein.png',0)
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('E:/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = detect_face(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
def adj_detect_face(img):
    
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = adj_detect_face(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('E:\haarcascade_eye.xml')
```

```python
def detect_eye(img):
    
    face_img = img.copy()
    
    eye_rects = eye_cascade.detectMultiScale(face_img)
    
    for(x,y,w,h) in eye_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
results = detect_eye(nadia)
plt.imshow(results,cmap='gray')
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow('Video Detect Face',frame)
    
    k = cv2.waitKey(1)
    if k==27:
        break
        
        cap.release()
        cv2.destroyAllWindows()
```
